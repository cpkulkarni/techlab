```python

# Python3 program to swap first
# and last element of a list

# Swap function
def swapList(newList, first_position, second_position):
    size = len(newList)

    # Swapping
    temp = newList[first_position]
    newList[0] = newList[second_position]
    newList[second_position] = temp

    return newList
#print("test")
    
# Driver code
newList = [12, 35, 9, 56, 24]

print(swapList(newList, 1, 2))
```

```python
def max_number(numList):
  max_no = max(numList)
  min_no = min(numList)
  return max_no, min_no

newList = [1,2,1414,2015,1141]
print("Max number is ",max_number(newList)[0], "min number is ", max_number(newList)[1] )
```

```python
# How to Remove Letters From a String in Python

def rev_string(iStr):

  revStr = ""
  # using split and join
  #revStr = revStr.join(iStr.split("@"))

  # using replace method
  revStr = iStr.replace("@","")

  return revStr

newStr = "me@office"

print(rev_string(newStr))
```
